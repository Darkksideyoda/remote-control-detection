```python
from keras.models import model_from_json
from PIL import Image
import numpy as np
import io
import matplotlib.pyplot as plt
from keras.models import load_model
from tensorflow.keras.preprocessing import image
```

```python
width = 150
height = 150

model = model_from_json(open("mmodel_weights.json","r").read())
model.load_weights("remoteControl.h5")

```

```python
img =  Image.open("data/collectiveData/notRemote_50.jpg")
plt.imshow(img)
# im = Image.frombytes("RGB",img.size,img.rgb)

im2 = np.array(img.convert("L").resize((width,height)))   

im2 = im2/255
plt.imshow(im2)

X = np.array([im2])
X = X.reshape(X.shape[0],width,height,1)
r = model.predict(X)
result = np.argmax(r)
print(result)
```
